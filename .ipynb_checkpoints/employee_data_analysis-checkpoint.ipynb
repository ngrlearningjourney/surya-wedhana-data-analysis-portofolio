# Question
1. How many employee of each degree that leave the company and how much is not?
2. How many employee that got recruited each year?
3. What is the average employee age from each education degree

```python
# How many employee of each degree that leave the company and how much is not?
leave_or_not_mapping = {
    1 : "Leave",
    0 : "Not Leave"
}

def replaceLeaveorNot(LeaveOrNot):
    return leave_or_not_mapping.get(LeaveOrNot)

employee_df['LeaveOrNot'] = employee_df['LeaveOrNot'].apply(replaceLeaveorNot)
employee_df.head(5)
```

```python
employee_df.groupby(by=["Education","LeaveOrNot"]).agg({
    "employee_id" : "count"
}).sort_values(by="employee_id", ascending=False)
```

Thorugh the code above, we know:

there are 1232 peoples that have bachelor degree not leaving the company, and also we have 328 Masters that not leave the company and last we have 116 PHD that haven't leave the company

```python
education_leaveornot_df = employee_df.groupby(by=["Education","LeaveOrNot"]).agg({
    "employee_id" : "count"
}).sort_values(by="employee_id", ascending=False).reset_index()
education_leaveornot_df.columns = ['Education','LeaveOrNot' ,'NumberofEmployee']
```

```python
sns.barplot(data = education_leaveornot_df, x='Education', y="NumberofEmployee", hue="LeaveOrNot", errorbar=None)
plt.title('Comparing Each Degree From Employee')
plt.ylabel('Number Of Employee')
plt.show()
```

```python
# How many employee that got recruited each year?
employee_df.groupby(by=['JoiningYear','Education']).agg({
    'employee_id' : 'count'
})
```

```python
joiningyear_education_df = employee_df.groupby(by=['JoiningYear','Education']).agg({
    'employee_id' : 'count'
}).reset_index()

joiningyear_education_df.rename(columns={
    "employee_id" : "NumberofEmployee"
}, inplace=True)

joiningyear_education_df
```

```python
bachelors_df = joiningyear_education_df[joiningyear_education_df['Education'] == 'Bachelors']
masters_df = joiningyear_education_df[joiningyear_education_df['Education'] == 'Masters']
phd_df = joiningyear_education_df[joiningyear_education_df['Education'] == 'PHD']

# plot lines
plt.plot(bachelors_df['JoiningYear'], bachelors_df['NumberofEmployee'], label = "Bachelors")
plt.plot(masters_df['JoiningYear'], masters_df['NumberofEmployee'], label = "Masters")
plt.plot(phd_df['JoiningYear'], phd_df['NumberofEmployee'], label = "PHD")
plt.legend()
plt.show()
```

# Classification

```python
employee_df = employee_df.replace(['Bachelors', 'Masters', 'PHD','Bangalore', 'Pune', 'New Delhi', 'Male','Female','No','Yes','Not Leave', 'Leave'],
                 [1,2,3,1,2,3,1,2,1,2,1,2])
```

```python
X = employee_df.iloc[:,:-1].values
y = employee_df.iloc[:,-1].values
# x is independent variable and y is dependent variable
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1)
# Split the dataset into training and testing sets
# X: Features (input data)
# y: Target (output labels)
# test_size: Proportion of the dataset to include in the test split (10% in this case)
# X_train, X_test: Training and testing features
# y_train, y_test: Training and testing labels
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
# StandardScaler standardizes the data by removing the mean and scaling to unit variance
sc = StandardScaler()

# Fit the scaler to the training data and transform it
# This calculates the mean and standard deviation from X_train and applies scaling
X_train = sc.fit_transform(X_train)

# Transform the test data using the same scaling parameters calculated from the training data
# to ensure consistency between training and test data scaling.
X_test = sc.transform(X_test)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

# Initialize the Logistic Regression classifier
# random_state: Ensures reproducibility by setting a fixed random seed
classifier = LogisticRegression(random_state=0)

# Fit the Logistic Regression model to the training data
# X_train: Training features
# y_train: Training labels
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred)
print("Confusion Matrix\n", cm)
print("accuracy score", accuracy_score(y_test,y_pred))
```
